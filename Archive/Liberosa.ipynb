```python
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "sample.wav"
```

```python
# signal, sr = librosa.load(file,sr=22050) #sr * T -> 22050 * duration
# librosa.display.waveshow(signal, sr=sr)4
time        = np.arange(0, 1, 0.01);
amplitude   = np.sin(2*np.pi*2*time)

plt.plot(time,amplitude)
plt.xlabel("Time")
plt.ylabel("Amplitude")

```

```python
fft = np.fft.fft(amplitude)
magnitude = np.abs(fft)/100
frequency = np.linspace(0,1,len(magnitude))
left_frequency = frequency[:int(len(frequency))]
left_magnitude = magnitude[:int(len(magnitude))]
plt.plot(left_frequency,left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```
