```python
# Contract a time domain signal
import numpy as np
import matplotlib.pyplot as plt

Fs = 100 # Sampling frequency
tstep = 1/Fs # sample time interval
f0 = 10 # signal frequency
N = int(Fs/f0)*3 # Number of samples of signal
t = np.linspace(0,(N-1)*tstep, N) # time steps
fstep = Fs / N # Frequency interval 
f = np.linspace(0, (N-1)*fstep, N)

y = 1 * np.sin(2*np.pi * f0 * t) + 2 * np.sin(2*np.pi * 23 * t) 



# Perform FFT
X = np.fft.fft(y)
X_mag = np.abs(X) / N

f_plot = f[0:int(N/2+1)]
f_mag_plot = 2 * X_mag[0:int(N/2+1)]
f_mag_plot[0] = f_mag_plot[0]/2
# Plot

fig, [ax1,ax2] = plt.subplots(nrows=2, ncols=1)
ax1.plot(t,y,'.-')
ax2.plot(f_plot,f_mag_plot,'.-')

plt.show()

```
