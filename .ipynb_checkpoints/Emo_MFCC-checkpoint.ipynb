```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio
import warnings
warnings.filterwarnings("ignore")
```

## Load the dataset

```python
paths = []
labels = []

for root, dirs, files in os.walk("./input/", topdown=False):
    for name in files:
        label = name.split('_')[-1].split('.')[0].lower()
        labels.append(label)
        paths.append(os.path.join(root,name))
        
print("Dataset Loaded")
```

```python
df = pd.DataFrame()
df['speech'] = paths
df['label'] =labels
```

```python
sns.countplot(df['label'])
```

```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion,size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()

def spectogram(data,sr,emotion):
    x = librosa.stft(data)
    xdb = librosa.amplitude_to_db(abs(x))
    plt.figure(figsize=(10,4))
    plt.title(emotion,size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time',y_axis='hz')
    plt.colorbar()
    
```

```python
emotions = ['neutral','angry']

for emotion in emotions:
    path = df['speech'][df['label'] == emotion]
    data, sampling_rate=librosa.load(path.iloc[1])
    waveplot(data, sampling_rate,emotion)
    spectogram(data,sampling_rate,emotion)
Audio(path.iloc[1])
```

## Feature Extraction

```python
def extract_mfcc(filename):
    y ,sr = librosa.load(filename, duration=3, offset=0.5)
    mfcc =np.mean(librosa.feature.mfcc(y=y,sr=sr,n_mfcc=40).T,axis=0)
    return mfcc
```

```python
extract_mfcc(df.loc[0,'speech'])
```

```python
X_mfcc = df['speech'][0:100].apply(lambda x: extract_mfcc(x))
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
```

## LSTM MODEL

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(40,1)),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax'),
])

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

## Train the model

```python
# y = np.asarray(y).astype(np.float32)
X = np.asarray(X).astype(np.float32)
history = model.fit(X,y,validation_split=0.2, epochs=1, batch_size=512,shuffle=True)
```
