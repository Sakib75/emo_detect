```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio
import warnings
warnings.filterwarnings("ignore")
```

## Load the dataset

```python
paths = []
labels = []

for root, dirs, files in os.walk("./TESS/", topdown=False):
    for name in files:
        label = name.split('_')[-1].split('.')[0].lower()
        labels.append(label)
        paths.append(os.path.join(root,name))
        
print(f"{len(labels)} DATA Loaded")
```

```python
df = pd.DataFrame()
df['speech'] = paths
df['label'] =labels
```

```python
sns.countplot(df['label'])
```

```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion,size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()

def spectogram(data,sr,emotion):
    x = librosa.stft(data)
    xdb = librosa.amplitude_to_db(abs(x))
    plt.figure(figsize=(10,4))
    plt.title(emotion,size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time',y_axis='hz')
    plt.colorbar()
    
```

```python
emotions = ['neutral','angry']

for emotion in emotions:
    path = df['speech'][df['label'] == emotion]
    data, sampling_rate=librosa.load(path.iloc[1])
    waveplot(data, sampling_rate,emotion)
    spectogram(data,sampling_rate,emotion)
Audio(path.iloc[1])
```

## Feature Extraction

```python
def extract_mfcc(filename):
    y ,sr = librosa.load(filename, duration=5, offset=0.5)
    mfcc =np.mean(librosa.feature.mfcc(y=y,sr=sr,n_mfcc=40).T,axis=0)
    return mfcc
```

## Extract mfcc 

```python
X_mfcc = df['speech'][0:10].apply(lambda x: extract_mfcc(x))
X_mfcc
```

## Expand dimension

```python
X = [ x for x in X_mfcc]

X= np.array(X)

print(X_mfcc.shape)
print(X_mfcc[0])
print(X.shape)
print(X[0])
```

```python

X=np.expand_dims(X, -1)
print(X.shape)
print(X[0])
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()

y = enc.fit_transform(df[['label']][0:10])

print(y.shape)
y
enc.categories_[0]
```

```python
y = y.toarray()
print(y.shape)
y
```

## LSTM MODEL

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
unq = len(df['label'].unique())
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(40,1)),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(unq, activation='softmax'),
])

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

## Train the model

```python
# y = np.asarray(y).astype(np.float32)
# X = np.asarray(X).astype(np.float32)
print(y)
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, shuffle= True)
history = model.fit(x=x_train, y=y_train, batch_size=64, epochs=100, validation_data=(x_valid, y_valid))
```

## Plot the result

```python
Y_pred = model.predict(X)

np.argmax(Y_pred)

y_pred_labels = [np.argmax(i) for i in Y_pred]
y_real_labels = [np.argmax(i) for i in y]
```

```python
import tensorflow as tf
cm = tf.math.confusion_matrix(labels=y_real_labels,predictions=y_pred_labels)
```

```python
import seaborn as sn
plt.figure(figsize= (10,7))
sn.heatmap(cm,annot=True,fmt='d',xticklabels=enc.categories_[0],yticklabels=enc.categories_[0])
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
